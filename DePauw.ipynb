```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
DePauw_game = pd.read_csv("Games 14 DePauw.xlsx - DePauw.csv")
```

```python
DePauw_full = DePauw_game[DePauw_game['Split Name']== 'game']
DePauw_full
```

```python
DePauw_volume_intensity = DePauw__full[["Player Name", "Minutes Played", "Distance (miles)", "Sprint Distance (yards)", "Power Plays", "Energy (kcal)", "Impacts", "Top Speed (mph)","Distance Per Min (yd/min)", "Power Score (w/kg)", "Player Load", "Work Ratio"]]
```

```python
DePauw_cluster = DePauw_volume_intensity.loc[:, DePauw_volume_intensity.columns != 'Player Name']
```

```python
DePauw_distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(DePauw_cluster)
    DePauw_distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, DePauw_distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method shwoing the optimal k')
plt.show()
```

```python
x_DePauw = DePauw_cluster.values # numpy array
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x_DePauw)
X_norm = pd.DataFrame(x_scaled)
```

```python
DePauw_pca = PCA(n_components = 2) # 2D PCA for the plot
reduced_DePauw = pd.DataFrame(DePauw_pca.fit_transform(X_norm))
```

```python
kmeans = KMeans(n_clusters=3)
# fit the input data
kmeans = kmeans.fit(reduced_DePauw)
# get the cluster labels
labels = kmeans.predict(reduced_DePauw)
# centroid values
centroid = kmeans.cluster_centers_
# cluster values
clusters = kmeans.labels_.tolist()
# adding names
DePauw_names = DePauw_full['Player Name']
```

```python
reduced_DePauw['cluster'] = clusters
reduced_DePauw['names'] = DePauw_names
reduced_DePauw.columns = ['x', 'y', 'cluster', 'names']
reduced_DePauw.head()
```

```python
%matplotlib inline
sns.set(style="white")
ax = sns.lmplot(x="x", y="y", hue='cluster', data = reduced_DePauw, legend= True,
fit_reg=False, height = 15, scatter_kws={"s": 250})
texts = []
for x, y, s in zip(reduced_DePauw.x, reduced_DePauw.y, reduced_DePauw.names):
    texts.append(plt.text(x, y, s))
ax.set(ylim=(-2, 2))
plt.tick_params(labelsize=15)
plt.xlabel(" DePauw PC 1", fontsize = 20)
plt.ylabel(" DePauw PC 2", fontsize = 20)
plt.show()
```
