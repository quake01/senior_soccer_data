```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
Trine_game = pd.read_csv("Game 8.xlsx - Trine.csv")
```

```python
Trine_game
```

```python
Trine_full = Trine_game[Trine_game['Split Name']== 'all']
Trine_full
```

```python
Trine_volume_intensity = Trine_full[["Player Name", "Minutes Played", "Distance (miles)", "Sprint Distance (yards)", "Power Plays", "Energy (kcal)", "Impacts", "Top Speed (mph)","Distance Per Min (yd/min)", "Power Score (w/kg)", "Player Load", "Work Ratio"]]
```

```python
Trine_cluster = Trine_volume_intensity.loc[:, Trine_volume_intensity.columns != 'Player Name']
```

```python
Trine_distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(Trine_cluster)
    Trine_distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, Trine_distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method shwoing the optimal k')
plt.show()
```

```python
x_Trine = Trine_cluster.values # numpy array
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x_Trine)
X_norm = pd.DataFrame(x_scaled)
```

```python
pca_Trine = PCA(n_components = 2) # 2D PCA for the plot
reduced_Trine= pd.DataFrame(pca_Trine.fit_transform(X_norm))
```

```python
kmeans = KMeans(n_clusters=3)
# fit the input data
kmeans = kmeans.fit(reduced_Trine)
# get the cluster labels
labels = kmeans.predict(reduced_Trine)
# centroid values
centroid = kmeans.cluster_centers_
# cluster values
clusters = kmeans.labels_.tolist()
# adding names
Trine_names = Trine_full['Player Name']
```

```python
reduced_Trine['cluster'] = clusters
reduced_Trine['names'] = Trine_names
reduced_Trine.columns = ['x', 'y', 'cluster', 'names']
reduced_Trine.head()
```

```python
%matplotlib inline
sns.set(style="white")
ax = sns.lmplot(x="x", y="y", hue='cluster', data = reduced_Trine, legend= True,
fit_reg=False, height = 15, scatter_kws={"s": 250})
texts = []
for x, y, s in zip(reduced_Trine.x, reduced_Trine.y, reduced_Trine.names):
    texts.append(plt.text(x, y, s))
ax.set(ylim=(-2, 2))
plt.tick_params(labelsize=15)
plt.xlabel(" Trine PC 1", fontsize = 20)
plt.ylabel(" Trine PC 2", fontsize = 20)
plt.show()
```
